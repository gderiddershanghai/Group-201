```python
def get_embedding(word):
    inputs = tokenizer(word, return_tensors="pt")
    outputs = model(**inputs)
    # Get the embedding for the [CLS] token, representing the whole word/sentence
    embedding = outputs.last_hidden_state[:, 0, :].squeeze().detach().numpy()
    return embedding
```

```python
# Apply the embedding function to each language column
df['Chinese_embedding'] = df['Chinese'].apply(get_embedding)
df['English_embedding'] = df['English'].apply(get_embedding)
df['French_embedding'] = df['French'].apply(get_embedding)
```

```python
embeddings = np.hstack(df[['English_embedding', 'Chinese_embedding','French_embedding']].values)
embeddings.shape
```

```python
embeddings_matrix = np.vstack(df[['Chinese_embedding', 'English_embedding', 'French_embedding']].apply(
    lambda row: np.hstack(row.values), axis=1).values)
```

```python
embeddings = np.vstack(embeddings)
embeddings.shape
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Reduce to 2 or 3 dimensions
pca = PCA(n_components=2)
pca_embeddings = pca.fit_transform(embeddings)


```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'embeddings' is a numpy array of shape (303, 768) after stacking
# PCA reduction
pca = PCA(n_components=2)
pca_embeddings = pca.fit_transform(embeddings)

# Create a DataFrame for easy plotting
pca_df = pd.DataFrame(pca_embeddings, columns=["PC1", "PC2"])

# Assuming 'df' has the columns 'English', 'Chinese', 'French' for labels
# Concatenate the labels into one column for easy annotation
pca_df["Label"] = df["English"].tolist() + df["Chinese"].tolist() + df["French"].tolist()

# Set the color palette to have as many colors as there are unique words
num_colors = len(pca_df)
palette = sns.color_palette("hsv", num_colors)

# Plot with Seaborn and annotate each point
plt.figure(figsize=(15, 10))
sns.scatterplot(data=pca_df, x="PC1", y="PC2", hue=pca_df.index, palette=palette, legend=None, s=100, alpha=0.8)

# Annotate each point with the word label
for i, label in enumerate(pca_df["Label"]):
    if i==0:
    plt.text(pca_df.loc[i, "PC1"], pca_df.loc[i, "PC2"], label, fontsize=9, ha='right')

plt.title("PCA of Multilingual Embeddings with Unique Colors and Annotations")
plt.xlabel("PC 1")
plt.ylabel("PC 2")
plt.show()
```
