```python
def get_embedding(word):
    inputs = tokenizer(word, return_tensors="pt")
    outputs = model(**inputs)
    # Get the embedding for the [CLS] token, representing the whole word/sentence
    embedding = outputs.last_hidden_state[:, 0, :].squeeze().detach().numpy()
    return embedding
```

```python
# Apply the embedding function to each language column
df['Chinese_embedding'] = df['Chinese'].apply(get_embedding)
df['English_embedding'] = df['English'].apply(get_embedding)
df['French_embedding'] = df['French'].apply(get_embedding)
```

```python
embeddings = np.hstack(df[['English_embedding', 'Chinese_embedding','French_embedding']].values)
embeddings.shape
```

```python
from matplotlib import rcParams
import matplotlib.patches as mpatches

rcParams['font.sans-serif'] = ['Noto Sans CJK SC']  # Use the Noto Sans CJK SC font for Chinese characters
rcParams['axes.unicode_minus'] = False  # Ensure minus signs appear correctly

# Define a large color palette
num_colors = len(color_indices)
color_palette = sns.color_palette("tab20", num_colors)  # Generates a wide range of colors

plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")

# Plot all points in gray for a neutral background
sns.scatterplot(data=pca_df, x="PC1", y="PC2", color="gray", alpha=0.5)

# Dictionary to store the English word and its associated color for the legend
legend_elements = {}

# Loop through each triplet to set color and annotate each point separately
for idx, (eng_idx, chn_idx, fr_idx) in enumerate(color_indices):
    color = color_palette[idx]  # Get the color from the expanded palette
    
    # Add English word and color to the legend dictionary
    legend_elements[pca_df.loc[eng_idx, "Label"]] = color
    
    # English annotation
    plt.scatter(pca_df.loc[eng_idx, "PC1"], pca_df.loc[eng_idx, "PC2"], color=color, s=100)
    plt.text(pca_df.loc[eng_idx, "PC1"] - 0.31, pca_df.loc[eng_idx, "PC2"], pca_df.loc[eng_idx, "Label"],
             fontsize=9, color=color, ha='right', va='center')
    
    # Chinese annotation
    plt.scatter(pca_df.loc[chn_idx, "PC1"], pca_df.loc[chn_idx, "PC2"], color=color, s=100)
    plt.text(pca_df.loc[chn_idx, "PC1"] - 0.31, pca_df.loc[chn_idx, "PC2"], pca_df.loc[chn_idx, "Label"],
             fontsize=9, color=color, ha='right', va='center')
    
    # French annotation
    plt.scatter(pca_df.loc[fr_idx, "PC1"], pca_df.loc[fr_idx, "PC2"], color=color, s=100)
    plt.text(pca_df.loc[fr_idx, "PC1"] - 0.31, pca_df.loc[fr_idx, "PC2"], pca_df.loc[fr_idx, "Label"],
             fontsize=9, color=color, ha='right', va='center')

# Create custom legend based on the legend_elements dictionary
legend_handles = [mpatches.Patch(color=color, label=label) for label, color in legend_elements.items()]
plt.legend(handles=legend_handles, title="English Labels", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.title("PCA of Multilingual BERT Embeddings")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```
